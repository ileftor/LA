__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$


Приравняем линейную комбинацию данных функций к нулевому вектору:
$$\alpha_{1}e^x+\alpha_{2}+\alpha_{3}(x+1)+\alpha_{4}(x+e^x)=0$$

Преобразуем:
$$(\alpha_{1}-\alpha_{4})e^x+(\alpha_{3}+\alpha_{4})x+\alpha_{2}+\alpha_{3}=0$$

Тогда, если предположить, что функции линейно независимы, можно сделать вывод, что все коэффициенты должны равняться нулю. Получаем систему уравнений:
$$\begin{cases}
 \alpha_{1}-\alpha_{4}=0,\\
 \alpha_{3}+\alpha_{4}=0,\\
 \alpha_{2}=0,\\
 \alpha_{3}=0,
\end{cases}$$
$$\begin{cases}
 \alpha_{1}=\alpha_{4},\\
 \alpha_{4}-\alpha_{4}=0,\\
 \alpha_{2}=0,\\
 \alpha_{3}=0,
\end{cases}$$
откуда следует, что $\alpha_{1}=\alpha_{2}=\alpha_{3}=\alpha_{4}=0$. Таким образом, функции линейно независимы.
